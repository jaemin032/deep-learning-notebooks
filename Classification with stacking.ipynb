```python
from sklearn.model_selection import cross_val_score
from sklearn.datasets import load_iris
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
```

```python
for k in iris:
    print(k)
```

```python
data = iris['data']
target = iris['target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size=0.33)
```

```python
from sklearn.linear_model import SGDClassifier
```

```python
linreg = SGDClassifier()
linreg.fit(x_train, y_train)
linreg.score(x_test, y_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ada = AdaBoostClassifier(n_estimators=10000)
ada.fit(x_train, y_train)
ada.score(x_test, y_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbr = GradientBoostingClassifier(n_estimators=10000)
gbr.fit(x_train, y_train)
gbr.score(x_test, y_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
```

```python
def build_classification_model():
    model = Sequential()
    model.add(Dense(8, input_shape=(4,)))
    model.add(Dropout(0.1))
    model.add(Dense(3, activation='softmax'))
    model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
neural_net = build_classification_model()
```

```python
neural_net.fit(x_train, y_train, epochs=500)
```

```python
loss, nn_accuracy = neural_net.evaluate(x_test, y_test)
```
